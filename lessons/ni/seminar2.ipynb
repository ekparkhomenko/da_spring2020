Сначала рассчитаем оценку среднего значения выборки ( среднюю массу в граммах для посылки)

```python
mean =y_summ/n
mean
```

Далее рассчитаем оценку общего ( суммарного значения) для генеральной совокупности

```python
Y_summ =mean*N
Y_summ
```

Посчитаем выборочную дисерсию

```python
var = (y_summ_2 -np.power(y_summ,2)/n)/(n-1)
var
```

И стандартное отклоненение

```python
std =np.sqrt(var)
std
```

Стандартная ошибка среднего (насколько оценка среднее значение отклоняется от среднего)

```python
std_mean_error =(std*np.sqrt(1-n/N))/np.sqrt(n)
std_mean_error
```

Стандартная ошибка суммарного значения (насколько оценка суммарного значение отклоняется от суммы ГС)

```python
std_summ_error =N*(std*np.sqrt(1-n/N))/np.sqrt(n)
std_summ_error
```

Для того, чтобы рассчитать доверительные интервалы нужно найти квантиль _t_ уровня  _а_  нормального распределения _N(0,1)_. В приведенной формуле, _а_ - вероятность ошибки. Если нам необходимы 80% доверительные границы, то альфа равна 20%. 

```python
from scipy.stats import norm
```

```python
alpha =1-0.95
```

```python
#Левая квантиль
t_2a=norm.ppf(alpha/2)
t_2a
```

```python
#Правая квантиль
t_1_2a=norm.ppf(1-alpha/2)
t_1_2a
```

```python
## Рассчитаем левую и правую границу доверительного интервала для среднего значения
```

```python
Y_mean_left =mean+t_2a*std_mean_error
Y_mean_right =mean+t_1_2a*std_mean_error
(Y_mean_left,Y_mean_right)
```
